```python
#page.find_all('div',class_='alphabetical-box')[0]  #at this class at attributes on 0
#page.find_all('div',class_='alphabetical-box')[0].find_all('a')   # just a tag on 0 index
#page.find_all('div',class_='alphabetical-box')[0].find_all('a')[0] # jus a tag of 0 indx of a
page.find_all('div',class_='alphabetical-box')[0].find_all('a')[0].attrs['href']   # getting link of the 0th index value
```
