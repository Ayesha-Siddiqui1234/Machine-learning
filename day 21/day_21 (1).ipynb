```python
#BAR PLOT (NUMERICAL AND CATEGORICAL)  # in bar plot we dont have size and style we have x y hue hue_order
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age']) #bivarient analysis its shows that in Pclass 3 younger and in Pclass 1 older age is travellig the middle line is the confidence support which shows that if there arr 5 popel so asking their age so their is more accuracy so least error line of confience suport is lower if line is greater so the confieence support is higher more chances of accuracy bcz more porple like 1000 so chnace of erorr are high high confidenc support high line
```

```python
#sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])   # multi varient analysis barchrt only 3 paraemter are used
#now i want to add the 4th parameter so which i have to use catplot
#x: Horizontal groups.

#y: Numerical value.

#hue: Color-coded sub-groups.

#col: Creates a new column of charts for another category.

#row: Creates a new row of charts for a fifth category.

sns.catplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'],col=titanic['Survived'],row=titanic['Embarked'])
```

```python
#DISTPLOT   is used for numerical colomn
sns.distplot(titanic[titanic['Survived'] == 0]['Age'],hist=False)  # the prople who died
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)    #graph shows the survival rate of child orange is higher than the died rate but in old age the surival age is lower
```
